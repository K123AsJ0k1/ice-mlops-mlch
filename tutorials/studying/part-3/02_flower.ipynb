## Why use Flower?

Flower was chosen for the following reasons:

- Most common monitoring application for Celery (mature)

- Easy to use and setup to gather Celery metrics (abstracted)

- Usable in any place where Python can be used (interoperable)

These enable us to use Flower to debug and observe containerized Celery applications in the same system.

## How to use Flower?

Assuming we have a running Redis container [(1)](#used-material-1), we can start the Flower monitor by setting up a venv |[(2)](#used-material-2),[(3)](#used-material-3),[(4)](#used-material-4)| and executing the [run](./celery/run_celery.py) file in the following way [(5)](#used-material-5):

```
cd flower
python3 -m venv flower_venv
source flower_venv/bin/activate
pip install -r packages.txt
python3 run_flower.py
```

We can now check the Celery instances and tasks at http://127.0.0.1:7501 by giving the username flower123 and password flower456. The relevant pages are:

- Workers = Shows registered Celery workers
- Tasks = Shows registered Celery tasks during Flower run

## Flower file structure

When you check the files inside the flower folder, you notice the following structure:

- [packages](./flower/packages.txt)
- [run_flower](./flower/run_flower.py)
- [setup_flower](./flower/setup_flower.py)

The purpose of this structure is to keep things as simple as possible, since the software only requires configuration to be useful.
