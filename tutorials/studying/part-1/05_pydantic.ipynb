## Why use Pydantic?

Pydantic was chosen for the following reasons:

- One of the most used validation tools in Python (mature)

- Simple syntax for creating complex models (abstracted)

- Usable in any place where Python can be used (interoperable)

These enable us to use Pydantic as a data modeling and validation tool to quickly design and test our automation objects.

## How to use Pydantic?

Assuming you already setup your venv environment [(1)](#used-material-1), let's start by creating a data model. Be aware that we will not go into the syntax details unless they are really important, because they are better explained in the Pydantic docs [(2)](#used-material-2). Create the following model by running the block:

```python
from pydantic import BaseModel, Field
from typing import List

class Venv(BaseModel):
    name: str = Field(alias = 'name')
    path: str = Field(alias = 'path')
    modules: List[str] = Field(default_factory = list, alias = 'modules')
    packages: List[str] = Field(default_factory = list, alias = 'packages')
```

We can now create equivalent dictionaries, such as:

```python
test_venv = {
    'name': 'tutorial-venv',
    'path': '/projappl/[CSC_PROJECT]',
    'modules': [
        'load pytorch/2.9'
    ],
    'packages': [
        'ray',
        'python-swiftclient',
        'torch',
        'torchmetrics',
        'qdrant-client'
    ]
}
```

We can validate, print, and format it in the following ways:

```python
validated_venv = Venv.model_validate(test_venv)
```

```python
print(validated_venv)
```

```python
validated_venv.model_dump()
```

We can create and validate nested models in the following way:

```python
class Configs(BaseModel):
    venv: Venv = Field(alias = 'venv')
```

```python
test_configs = {
    'venv': test_venv
}
```

```python
validated_configs = Configs.model_validate(test_configs)
```

```python
print(validated_configs)
```

```python
validated_configs.model_dump()
```

This shows the advantage of validators such as Pydantic. Instead of creating a large dictionary containing both venv and configs, which we check ourselves, we split them into smaller parts stored in separate files. This kind of division by creating separate abstraction layers will be the main tool for ensuring that integration users can find what they need, modify it, and test it as its own unit.
