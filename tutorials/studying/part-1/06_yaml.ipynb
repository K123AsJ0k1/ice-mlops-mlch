## Why use YAML?

YAML was chosen for the following reasons:

- One of the most widespread formats in computer science (mature)

- Simple way to write and read complex data structures (abstraction)

- Can be used anywhere that enables storing and parsing text (interoperability)

These enable us to use YAML as a data serialization language to easily construct, edit, and verify complex automation configurations.
